#Trabalhando com número variado de argumentos em Funções Python

```python
#Quando não souber quantos argumentos passar para função

#Argumentos de tamanho variável(*args)

def soma_numeros(*args):
    """Soma um número variável de argumentos"""
    total = 0
    for numero in args:
        total += numero
    return total
```

```python
print(f"Soma dos números: {soma_numeros(1,2,3,4,5)}")
```

```python
print(f"Soma dos Números: {soma_numeros(1,2,3)}")
```

```python
#Quando quiser trabalhar com dicionário ou outra estrutura de dados parecida com dicionário

#Argumentos de tamanho variável(**kwargs)
def exibe_info_pessoa(**kwargs):
    """Exibe informações passadas como pares chave-valor."""
    print("\nInformações da Pessoa: \n")

    for chave, valor in kwargs.items():
        print(f"-{chave}: {valor}")
```

```python
#Chamando a função

exibe_info_pessoa(nome = "Carla",
                  profissao = "Engenheira de Dados",
                  hobby = "Leitura")
```

```python
#Chamando a função

exibe_info_pessoa(nome= "Bob",
                  profissao= "Cientista de Dados")
```

#Funções Anônimas

```python
#Uma função lambda que dobra um número

dobro = lambda x:x * 2

#Print

print(f"O dobro de 7 é : {dobro(7)}")
```

```python
#Lista de númetos

numeros = [1,2,3,4,5]

#lambda retorna o quadrado de cada elemento

quadrados = list(map(lambda x : x **2, numeros))

print(quadrados)
```

```python
#Lista de números

numeros = [1,2,3,4,5,6]


#Primeiro calculamos os quadrados com map + lambda

quadrados = list(map(lambda x:x ** 2, numeros))


#Agora filtramos apenas os pares com filter + lambda

quadrados_pares = list(filter(lambda x:x % 2 == 0, quadrados))

print("Quadrados: ", quadrados)
print("Quadrados pares: " , quadrados_pares)
```
