#Escopo de Variáveis

```python
#Variáveis Globais
saudacao = "Olá Mundo"
nome = "Aluno"

#Função
def minha_funcao():

    #Variável Local
    nome = "Ana"
    print(f"\n Dentro da função: {nome}")
    print(f"\n Acessando a variável global dentro da função: {saudacao}")

minha_funcao()

print(f"\n Fora da função: {saudacao}")
print(f"\n Fora da função: {nome}")
```

#Manipulação de Strings

```python
#Define uma variável do tipo string

frase = " Aprender Python é muito divertido! "
```

```python
#Concatenação

nome = "Maria"
saudacao = "Olá " + nome + "!"
print(f"{saudacao}")
```

```python
#Tamanho da String

print(f"Tamanho da frase: {len(frase)}")
```

```python
#Maiúsculas
print(f"Maiúsculas: {frase.upper()}")

#Minúsculas
print(f"Minúsculas: {frase.lower()}")
```

```python
#Remover espaços do início e do fim

frase_sem_espacos = frase.strip()
print(f"Frase sem espaços: {frase_sem_espacos}")
```

```python
#Substituir Texto

print(f"Substituindo 'divertido' por 'legal': {frase.replace('divertido', 'legal')} ")
```

```python
#Fatiamento(Slicing)- Acessando partes de uma string

#Índice em Python começa em 0

print(frase_sem_espacos)
print(f"O primeiro caractere: {frase_sem_espacos[0]}")
print(f"A palavra Python: {frase_sem_espacos[9:15]}")   #Do índice 9 ao 14
```

#Estrutura de Dados - Listas

```python
#Criando uma lista

frutas = ["maçã", "banana", "laranja", "abacaxi"]
print(f"Lista de frutas: {frutas}")
```

```python
type(frutas)
```

```python
#Acessando um item pelo índice

print(f"A primeira fruta é : {frutas[0]} ")          #buscando um valor no índice 0
print(f"A última fruta é: {frutas[-1]}")             #-1 (vai até o final da lista e busca o último elemento)
```

```python
#Adicionar um item ao final da lista

frutas.append("uva")
print(f"Lista após adicionar 'uva': {frutas}")
```

```python
#Removendo um item

frutas.remove("laranja")
print(f"Lista após remover 'laranja': {frutas} ")
```

```python
#Modificando um item

frutas[0] = "morango"
print(f"Lista após modificar o primeiro item: {frutas}")
```

```python
#Verificando o tamanho da lista

print(f" A lista tem {len(frutas)} frutas")
```

```python
#Deletando a lista

del frutas
```

#Estrutura de Dados - Dicionários

```python
#Criando um dicionário de informações de um aluno

aluno = {
    "nome": "Bob",
    "idade": 22,
    "curso": "Data Science",
    "aluno_ativo" : True
}
print(f"Dicionário do aluno: {aluno}")
```

```python
type(aluno)
```

```python
#Acessando um valor pela sua chave

print(f"Nome do aluno: {aluno['nome']}")
print(f"Curso: {aluno.get('curso')}")       #.get() é uma forma segura de acessar chaves
```

```python
#Adicionando um novo par chave-valor

aluno["cidade"] = "Recife"
print(f"Dicionário atualizado:\n {aluno}")
```

```python
#Modificar um valor existente

aluno["idade"] = 23
print(f"Idade atualizada: {aluno['idade']}")
```

```python
#Removendo um par chave-valor

del aluno["aluno_ativo"]
print(f"Dicionário após remover 'aluno_ativo':\n  {aluno}")
```

#Estrutura de Dados - Conjuntos(Sets)

```python
#Criando um conjunto

numeros = {1,2,3,4,2,3,5}
print(f"Conjunto de números (sem duplicatas): {numeros}")
```

```python
type(numeros)
```

```python
#Adicionando um item

numeros.add(6)
print(f"Após adicionar o valor 6 :  {numeros}")
```

```python
#Removendo um item

numeros.remove(2)
print(f"Após remover o valor 2:  {numeros}")
```

```python
#Operações de conjunto

conjunto_a = {1,2,3,4}
conjunto_b = {3,4,5,6}
print(f"Conjunto a: {conjunto_a}")
print(f"Conjunto b: {conjunto_b}")
```

```python
#União

uniao = conjunto_a.union(conjunto_b)
print(f"A união de A e B é:  {uniao}")
```

```python
#Interseção

intersecao = conjunto_a.intersection(conjunto_b)
print(f"A interseção de A e B é: {intersecao}")
```

#Conversão entre tipos de dados - Type Casting

```python
#Convertendo de String para número Integer

numero_em_texto = "123"
numero_inteiro = int(numero_em_texto)
print(f"String '{numero_em_texto}' para Inteiro:  {numero_inteiro}, Tipo: {type(numero_inteiro)}")
```

```python
#Convertendo de String para Float

numero_decimal_em_texto = "45.67"
numero_float = float(numero_decimal_em_texto)
print(f"String '{numero_decimal_em_texto}' para Float: {numero_float}, Tipo {type(numero_float)}")
```

```python
#Convertendo de Número para String

idade = 25
idade_em_texto = str(idade)
print(f"Inteiro {idade} para String: '{idade_em_texto}', Tipo: {type(idade_em_texto)}")
```

```python
#Convertendo entre estruturas de dados

lista_com_duplicatas = [1,2,2,3,4,4,4,5]      #lista
conjunto_sem_duplicatas = set(lista_com_duplicatas)    #conjunto
lista_sem_duplicatas = list(conjunto_sem_duplicatas)  #lista

print(f"\n Lista Original: {lista_com_duplicatas}")
print(f"\n Convertida para Conjunto: {conjunto_sem_duplicatas}")
print(f"\n Convertida de volta para Lista: {lista_sem_duplicatas}")
```
